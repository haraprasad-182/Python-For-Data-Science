```python
import numpy as np
import pandas as pd
```

```python

# First DataFrame: Marks
df1 = pd.DataFrame({
    'Name': ['A', 'A', 'C', 'D'],
    'Marks': [10, 20, 30, 40]
})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'Name': ['A', 'A', 'A', 'E', 'F'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male']
})
```

```python
df2
```

```python

pd.merge(df1,df2,on=['Name'],how = 'left')
```

```python
pd.merge(df1,df2,on=['Name'],how = 'right')
```

```python
pd.merge(df1,df2,on=['Name'],how = 'outer')
```

```python
pd.merge(df1,df2,on=['Name'],how = 'inner')
```

```python
# to get cross joint
df1['constant']=1
df2['constant']=1
```

```python
df1
```

```python
df2
```

```python

pd.merge(df1,df2,on='constant',how='inner')
```
